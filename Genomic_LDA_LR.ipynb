```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
dataset = pd.read_csv("GenomicData.csv")
dataset
```

```python
dataset = dataset.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])
dataset
```

```python
np.random.seed(42)
```

```python
X = dataset.drop(columns=['ClassType'])
y = dataset["ClassType"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
imputer = SimpleImputer(strategy='median')
imputer.fit(X_train)
```

```python
X_train_imp = imputer.transform(X_train)
X_test_imp = imputer.transform(X_test)
```

```python
scaler = StandardScaler()
scaler.fit(X_train_imp)
X_train_sc = scaler.transform(X_train_imp)
X_test_sc = scaler.transform(X_test_imp)
```

```python
LR1 = LogisticRegression(solver='liblinear', C=0.05).fit(X_train_sc, y_train)
LR2 = LogisticRegression(solver='liblinear', C=10).fit(X_train_sc, y_train)
```

```python
print(LR1.score(X_test_sc, y_test))
print(LR2.score(X_test_sc, y_test))
```

```python
LDA = LinearDiscriminantAnalysis().fit(X_train_sc, y_train)
```

```python
print(LDA.score(X_test_sc, y_test))
```

```python
print(np.count_nonzero(LR1.coef_))
print(np.count_nonzero(LR2.coef_))
print(np.count_nonzero(LDA.coef_))
```
